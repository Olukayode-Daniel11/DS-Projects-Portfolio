<p style="margin-top:0; margin-bottom:14px;">
Cardiovascular diseases (CVDs) are the leading cause of death globally, responsible for an estimated <strong>17.9 million deaths annually</strong>, accounting for <strong>31% of all global mortality</strong>. Four out of five of these deaths result from heart attacks and strokes, with one-third occurring prematurely in individuals under 70 years of age. Heart failure, a major consequence of CVDs, places immense pressure on healthcare systems and reduces quality of life worldwide. Early detection of individuals at high cardiovascular risk—especially those with <strong>hypertension, diabetes, hyperlipidaemia</strong>, or pre-existing heart disease—is therefore essential to reduce mortality and improve outcomes. However, traditional diagnostic methods often require extensive medical evaluations that may not always be accessible or affordable, highlighting the need for data-driven, machine learning-based predictive solutions.
</p>

<h3><b>Clinical Justification for Retaining Outliers</b></h3>
Outliers observed in <b>serum cholesterol</b>, <b>resting blood pressure</b>, and <b>ST-depression</b> were retained because, in clinical datasets, extreme values often represent severe pathological conditions rather than measurement errors. Removing such observations would bias the model toward mild cases and reduce real-world diagnostic performance. Values were therefore evaluated against physiologically plausible human ranges and kept if medically realistic, ensuring the model learns from the full spectrum of cardiovascular risk.

```python
#CORRELATION HEATMAP -
# TO CHECK FOR MULTICOLLINEARITY AMONGST THE NUMERIC INPUT FEATURES(only the numeric columns bcoz Correlation Heatmap only 
# deals with Numerical Values)
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Remove target variable for multicollinearity check
numeric_df = numeric_df.drop(columns=['heart_disease_present'])

plt.figure(figsize=(12,8))
corr = numeric_df.corr()

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            fmt='.2f',
            center=0,
            linewidths=0.5)

plt.title("Correlation Heatmap of Predictor Clinical Features")
plt.show()
```

To **reduce multicollinearity**, features with correlation coefficients above 0.6 were examined for redundancy. A strong association (r = 0.62) was observed between ST segment slope and exercise-induced ST depression, indicating overlapping physiological information. To maintain predictor independence, **ST depression (oldpeak) was retained** due to its clearer clinical interpretability, while **ST slope was excluded**. The final selected features — age, maximum heart rate achieved, number of major vessels, exercise-induced angina, and ST depression — demonstrate moderate or low intercorrelations, ensuring minimal multicollinearity while preserving strong cardiovascular relevance.

However, for **LVH (2)**, the number of patients with heart disease (around high-40s) is slightly higher than those without (mid-40s), indicating that abnormal ECG findings consistent with LVH are more frequently observed among patients with heart disease. Overall, the plot suggests that abnormal resting ECG findings, particularly LVH are associated with higher likelihood of heart disease in this dataset, while a normal ECG is more commonly seen in patients without heart disease.

Bivariate analysis evaluates the relationship between each feature and heart disease independently, highlighting variables that show strong individual associations with the target outcome. In contrast, logistic regression evaluates the independent predictive contribution of each variable while controlling for the influence of other variables in the dataset. As a result, some features that appear strongly associated with heart disease in isolation may lose statistical significance when overlapping clinical information is accounted for. This explains why variables such as sex emerged as statistically significant predictors in the logistic regression model, while other features such as thalassemia showed reduced influence after controlling for related cardiac stress indicators.

```python
#MODEL BUILDING

import statsmodels.api as sm

#Define Target and Features
X = df.drop("heart_disease_present", axis=1)
y = df["heart_disease_present"]
```

```python
# Convert target
y = pd.to_numeric(y)
```

```python
#Covert categorical columns properly
X = pd.get_dummies(X, drop_first=True)
```

```python
#Convert bool to int
bool_cols = X.select_dtypes(include=['bool']).columns
X[bool_cols] = X[bool_cols].astype(int)
```

```python
#Force Float
X = X.astype(float)
```

```python
#Add constant for intercept
X = sm.add_constant(X)
```

```python
#fit Logistic Regression Model
model = sm.Logit(y, X).fit()

print(model.summary())
```

```python
#PLOT ODDS RATIO TABLE

# Extract coefficients
coef = model.params

# Convert to odds ratios
odds_ratios = np.exp(coef)

# Confidence intervals
conf = model.conf_int()
conf.columns = ['Lower CI', 'Upper CI']

# Convert CI to odds ratio scale
conf['Lower OR'] = np.exp(conf['Lower CI'])
conf['Upper OR'] = np.exp(conf['Upper CI'])

# Combine everything
odds_table = pd.DataFrame({
    "Coefficient": coef,
    "Odds Ratio": odds_ratios,
    "P-value": model.pvalues,
    "Lower OR": conf["Lower OR"],
    "Upper OR": conf["Upper OR"]
})

odds_table.sort_values("P-value")
```

<h3><b>Odds Ratio Analysis of Key Predictors</h3></b>

The odds ratio analysis highlights several strong predictors of heart disease in the model. The **number of major vessels** detected through fluoroscopy emerged as one of the most influential factors, with an **odds ratio of approximately 2.78**, indicating that each additional affected vessel nearly triples the likelihood of heart disease. **Chest pain type** also showed a meaningful effect **(OR ≈ 1.93)**, suggesting that certain chest pain categories are associated with a higher probability of heart disease. Similarly, **ST depression during exercise (oldpeak)** demonstrated a significant relationship with disease presence **(OR ≈ 1.85)**, reflecting the importance of stress-induced cardiac abnormalities. Finally, **Sex** exhibited the strongest effect among the predictors **(OR ≈ 4.13)**, implying that the odds of heart disease differ substantially between males and females in the dataset. Overall, these findings emphasize that structural coronary abnormalities and stress-related cardiac indicators play a critical role in predicting heart disease risk.

```python
# NOTE: Odds Ratio Analysis gives a clinical interpretation of the coefficients values converted to odds ratio -
# by communicating what the odd ratio means from a medical stand point.

# NOTE: Another thing to note about "Odds Ratio" is that the result it gives naturally places the Top significant -
# features that influences the Target(y)data on top - in decsending order (From Highest to Lowest)
```

```python
y_pred_prob = model.predict(X)
y_pred_prob
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

fpr, tpr, thresholds = roc_curve(y, y_pred_prob)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label="Logistic Regression Model")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# COMPUTE ROC & AUC SCORE

from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Predicted probabilities
y_pred_prob = model.predict(X)

# AUC score
auc = roc_auc_score(y, y_pred_prob)

print("AUC Score:", auc)
```

<h3><b>ROC_AUC Analysis</h3></b>

The ROC analysis produced an Area Under the Curve **(AUC) score of 0.92**, indicating excellent model performance. An AUC value of 0.92 implies that the logistic regression model has a 92% probability of correctly distinguishing between patients with heart disease and those without the condition. This high score demonstrates strong discriminatory capability and suggests that the selected clinical predictors provide meaningful signals for identifying heart disease risk. Overall, the model significantly outperforms random classification and shows strong potential for predictive decision support in cardiovascular risk assessment.

```python
# CONFUSION MATRIX

from sklearn.metrics import confusion_matrix

y_pred = (y_pred_prob >= 0.5).astype(int)

cm = confusion_matrix(y, y_pred)
print(cm)
```

<H3><b> Confusion Matrix Insight</b></H3>

The confusion matrix provides a detailed evaluation of the classification performance of the logistic regression model. Out of <b>180 total observations</b>, the model correctly identified 87 patients without heart disease (true negatives) and successfully detected 65 patients with heart disease (true positives). However, the model produced 13 false positive predictions where healthy patients were incorrectly classified as having heart disease, and 15 false negative cases where patients with heart disease were incorrectly predicted as healthy. Overall, the model achieved a total of <b>152 correct predictions</b> and <b>28 incorrect predictions</b>. These results indicate that the model demonstrates strong predictive capability with a high number of correct classifications relative to misclassifications. The relatively low number of false positives and false negatives further supports the effectiveness of the model in distinguishing between patients with and without heart disease.

```python
# CLASSIFICATION REPORT

from sklearn.metrics import classification_report

print(classification_report(y, y_pred))
```
